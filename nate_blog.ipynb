```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
import urllib.parse
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q={}'

query = '빅데이터'
encoding = urllib.parse.quote(query)
url = url.format(encoding)

response = requests.get(url)
response.status_code

dom = BeautifulSoup(response.content, "html.parser")

post_elements = dom.select("a.f_link_b")
for l in post_elements:
    print(l.text, l.get("href"))
```

```python
dom = BeautifulSoup(response.content, "html.parser")
dom
```

```python
post_elements = dom.select("a.f_link_b")
post_elements
```

```python
for i in range(len(post_elements)):
    print (post_elements[i].text, post_elements[i].get("href"))
```

```python
df = pd.DataFrame(columns=["url", "title"])
for i in range(len(post_elements)):
    title = post_elements[i].text 
    url = post_elements[i].get("href")
    
    df.loc[len(df)] = [url, title]

df
```

```python
df.to_csv("nate_blog.csv")
```
